# 1. Bar(막대 그래프)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(x, x*2)
```

```python
np.random.seed(20221205)

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]

fig, ax = plt.subplots()
x_ax = np.arange(3)

# Draw a stacked bar graph starting from the bottom
for i in x_ax:
    ax.bar(x_ax,
           data[i],
           bottom = np.sum(data[:i], axis = 0))

# X-axis labeling
ax.set_xticks(x_ax)
ax.set_xticklabels(["A", "B", "C"])
```

# 2. Histogram

```python
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins = 50)  # bins = number of bars to display the histogram
```
