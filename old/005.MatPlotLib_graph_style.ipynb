# 1. Line graph(선 그래프)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
x = np.arange(15)
y = x ** 2
ax.plot(
    x, y,
    linestyle = ":",
    marker = "*",
    color = "#34d8eb"
)
```

# 2. Line Style

```python
x2 = np.arange(10)
y2 = x2
fig, ax = plt.subplots()
ax.plot(x2, y2, linestyle = "-")  # Solid
ax.plot(x2, y2 + 2, linestyle = "--")  # Dashed
ax.plot(x2, y2 + 4, linestyle = "-.")  # Dash-dotted
ax.plot(x2, y2 + 6, linestyle = ":")  # Dotted
```

# 3. Color

```python
x3 = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x3, x3, color="r")
ax.plot(x3, x3+2, color="yellow")
ax.plot(x3, x3+4, color="0.8")
ax.plot(x3, x3+6, color="#34d8eb")
```

# 4. Markers

```python
fig, ax = plt.subplots()
ax.plot(x3, x3, marker=".")  # Dot
ax.plot(x3, x3+2, marker="o")  # Circle
ax.plot(x3, x3+4, marker="v")  # Reversed triangle
ax.plot(x3, x3+6, marker="s")  # Square
ax.plot(x3, x3+8, marker="*")  # Asterisk
```

# 5. Axis boundary adjustment(축 경계 조정)

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)
ax.set_ylim(-1.5, 1.5)
```

# 6. Legend(범례)

```python
x = np.arange(15)
fig, ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc='upper right',
          shadow=True,
          fancybox=True,
          borderpad=2)
```

# 7. Other Options

### Line Styles

|Sign|Meaning|
|:---:|:----------:|
|-|Line|
|–-|Dash|
|-.|Dash-dotted|
|:|Dotted|

<br>

### Markers

|Sign|Meaning|Sign|Meaning|
|:---:|:---:|:---:|:---:|
|.|Dot|,|Pixel|
|o|Circle|s|Square|
|v, <, ^, >|Triangle|1, 2, 3, 4|Small triangle|
|p|Pentagon|H, h|Hexagon|


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(10)
fig, ax = plt.subplots()

ax.plot(
    x, x, label='y=x',
    linestyle='-',  # Line
    marker='.',  # Dotted
    color='blue'
)

ax.plot(
    x, x**2, label='y=x^2',
    linestyle='-.',  # Dash-dotted
    marker=',',  # Pixel
    color='red'
)

ax.set_xlabel("x")
ax.set_ylabel("y")
```
