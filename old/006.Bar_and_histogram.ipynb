# 1. Bar(막대 그래프)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Font Setting
import matplotlib as mpl

plt.rc("font", family = "AppleGothic")  # Mac 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지
```

```python
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(x, x*2)
```

```python
np.random.seed(20221205)

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]

fig, ax = plt.subplots()
x_ax = np.arange(3)

# Draw a stacked bar graph starting from the bottom
for i in x_ax:
    ax.bar(x_ax,
           data[i],
           bottom = np.sum(data[:i], axis = 0))

# X-axis labeling
ax.set_xticks(x_ax)
ax.set_xticklabels(["A", "B", "C"])
```

# 2. Histogram

```python
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins = 50)  # bins = number of bars to display the histogram
```

# 3. Sample Code

```python
# Data set
x = np.array(["축구", "농구", "야구", "배구", "탁구"])
y = np.array([13, 10, 17, 8, 7])
z = np.random.randn(1000)
np.random.seed(20221206)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Bar graph
axes[0].bar(x, y)

# Histogram
axes[1].hist(z, bins = 200)
```
