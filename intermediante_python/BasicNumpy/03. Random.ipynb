```python
import numpy as np

# Set the seed
np.random.seed(123)

# Initialize random_walk
random_walk = [0]

# Iterate 100 times
for step in range(100):
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1, 7)

    # Determine next step
    if dice <= 2:
        step -= 1
    elif dice <= 5:
        step += 1
    else:
        step += np.random.randint(1, 7)

    # Append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

```python
import numpy as np

np.random.seed(123)
random_walk = [0]

for step in range(100):
    step = random_walk[-1]
    dice = np.random.randint(1, 7)

    # Use max() to make sure step can't go below 0
    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step += 1
    else:
        step += np.random.randint(1, 7)

    # Append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

### Visualize above: visualization of the walk

```python
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

### Distribution of above

```python
# Initialize all_walks
all_walks = []

# Simulate random walk 100 times
for i in range(10):
    random_walk = [0]
    
    for j in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1, 7)
        
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

```python
# Visualization
# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()
```

```python
# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
# The purpose of transposing is to convert the rows of np_all_walks into columns.
# Every row in np_all_walks represents the position after 1 throw for the 10 random walks.
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

### Implement clumsiness
Let's consider that we have 0.1% chance of falling down.

Create a random float between 0 and 1.
If this value is less than or equal to 0.001, reset the step to 0.

```python
# Change the range() function to performe 250 times iteration
all_walks = []

for i in range(250):
    random_walk = [0]

    for j in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1, 7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)

        # Implement clumsiness
        if np.random.rand() <= 0.001:
            step = 0

        random_walk.append(step)

    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
# Select last row from np_aw_t: ends
ends = np_aw_t[-1, :]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```
