# 도수분포표(Frequency table) 출력

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Font Setting
import matplotlib as mpl

plt.rc("font", family = "AppleGothic")  # Mac 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지
```

```python
# drink data
drink = pd.read_csv("drink.csv")

# Calculate the frequency of each value
# .value_counts(): pandas의 series 객체에서 사용 가능한 함수로, 해당 series 객체의 값들의 빈도수를 계산해준다.
drink_freq = drink[drink["Attend"] == 1]["Name"].value_counts()

print("도수분포표")
print(drink_freq)
```

# Bar graph 출력

```python
# labes: drink_freq의 index
# ratio: drink_freq의 value
labels = drink_freq.index
ratio = drink_freq.values

fig, ax = plt.subplots()

plt.bar(labels, ratio)
plt.show()
```
